```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.examples.tutorials.mnist import input_data
import sys,os;
%matplotlib inline
sys.path.append("../");
from tf_data import *
```

```python
train_dir = '../data/MNIST/';
mnist = input_data.read_data_sets(train_dir, one_hot=True);
noisy_mnist = MNISTData.NoisyImages(train_dir, stddev=128);
```

```python
[n_x, n_y] = noisy_mnist.train.next_batch(batch_size=2);
```

```python
# Noisy image from our helper class
for im in n_x:
    plt.figure(figsize=(5,5));
    plt.imshow(np.reshape(im,newshape=(28,28)),cmap='gray');
```

```python
[m_x, m_y] = mnist.train.next_batch(batch_size=2);
```

```python
# Added noise to original images
for im in m_x:
    plt.figure(figsize=(10,10));
    plt.subplot(2,2,1);
    origin = np.array(255*im,dtype=np.uint8);
    plt.title("original")
    plt.imshow(np.reshape(origin,newshape=(28,28)),cmap='gray');
    plt.subplot(2,2,2);
    noisy = np.array(np.clip(255*im + 128 * np.random.randn(im.shape[0]),0,255),dtype=np.uint8);
    plt.title("noisy")
    plt.imshow(np.reshape(noisy,newshape=(28,28)),cmap='gray');
```
